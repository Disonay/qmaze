```python
import random
```

```python
plt.rcParams["figure.figsize"] = (32, 36)
plt.rcParams['font.size'] = 8
plt.rcParams["axes.grid"] = False
```

```python
free_cells = []
for i in range(maze.shape[0]):
    for ii in range(maze.shape[1]):
        if maze[i, ii] != 0:
            free_cells.append((i, ii))
```

```python
model = keras.models.load_model(model_path)
fig, ax = plt.subplots(1, 6)
for i in range(6):
    copy_maze = maze.copy()
    agent = random.choice(free_cells)
    free_cells.remove(agent)
    rlmaze = RLMaze(copy_maze, agent)
    actions = play_game(model, rlmaze, agent)
    for visit in actions:
        copy_maze[visit] = 0.5
    ax[i].matshow(copy_maze)
fig.tight_layout()
```
